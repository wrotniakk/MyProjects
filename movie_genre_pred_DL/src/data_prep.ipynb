```python
from ast import literal_eval
```

```python
def gender_mining(row):

    row = literal_eval(row)

    genres = []
    for i in row:
        genres.append(i["name"])

    return str(genres)
```

```python
movies["genre_list"] = movies["genres"].apply(gender_mining)
```

```python
movies.index=movies.title.values
```

```python
movies = movies[["genre_list","overview"]]
movies.dropna(inplace=True)
movies = movies[movies["genre_list"]!="[]"]
movies.info()
```

```python
movies.head()
```

## Narrowing down to only couple of most distinguishable genres

```python
genres = ['Fantasy','Horror','Romance','Action','Family']
```

```python
def filter(genre_list):

  gens = literal_eval(genre_list)

  return set(gens).issubset(set(genres))
```

```python
movies["mask"] = movies["genre_list"].apply(filter)
```

```python
movies.head()
```

```python
movies_narrowed = movies[movies["mask"]]
movies_narrowed.head()
```

```python
movies_narrowed = movies_narrowed[["genre_list","overview"]]
```

```python
movies_narrowed.to_csv("data/movies_genre&overview.csv")
```
